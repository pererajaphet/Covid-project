```python

import os
import librosa
import librosa.display
import IPython.display as ipd
import numpy as np
import matplotlib.pyplot as plt
```

```python
scale_file ='Downloads/hello/touxn.wav'
```

```python
ipd.Audio(scale_file)
```

```python
# load audio files with librosa
# the method load gives us it self the signal which I called here scale et le sample rate of the signal
#what is sample_rate in audio ? 
#it defines how many times per second a sound is sampled
#here Im getting approximattly (22KHz )

scale, sr = librosa.load(scale_file)
```

```python
#Extracting Short-Time Fourier Transform
```

```python
FRAME_SIZE = 2048
HOP_SIZE = 512
```

```python
S_scale = librosa.stft(scale, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)
```

```python
S_scale.shape
```

```python
type(S_scale[0][0])
```

```python
#Calculating the spectrogram
```

```python
Y_scale = np.abs(S_scale) ** 2
```

```python
Y_scale.mean()
```

```python
type(Y_scale[0][0])
```

```python
#Visualizing the spectrogram
```

```python
def plot_spectrogram(Y, sr, hop_length, y_axis="linear"):
    plt.figure(figsize=(25, 10))
    librosa.display.specshow(Y, 
                             sr=sr, 
                             hop_length=hop_length, 
                             x_axis="time", 
                             y_axis=y_axis)
    plt.colorbar(format="%+2.f")
```

```python
plot_spectrogram(Y_scale, sr, HOP_SIZE)
```

```python
#Log-Amplitude Spectrogram
```

```python
Y_log_scale = librosa.power_to_db(Y_scale)
plot_spectrogram(Y_log_scale, sr, HOP_SIZE)

Y_log_scale.mean()
```

```python
#Log-Frequency Spectrogram
```

```python
plot_spectrogram(Y_log_scale, sr, HOP_SIZE, y_axis="log")
```
